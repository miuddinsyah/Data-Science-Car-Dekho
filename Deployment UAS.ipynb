```python
import tkinter as tk
from tkinter import scrolledtext, Toplevel, Label, Entry, Button, Radiobutton, StringVar
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, classification_report, silhouette_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk
import random

# Load dataset
data = pd.read_csv('Dataset Viechle\Car details v3.csv')

# Preprocessing
data = data.dropna()  # Remove missing values
data = data.drop_duplicates(subset=['name', 'year', 'selling_price', 'km_driven', 'seller_type', 'mileage'], keep='first')

data['mileage'] = data['mileage'].str.extract(r'(\d+\.\d+)').astype(float)
data['engine'] = data['engine'].str.extract(r'(\d+)').astype(float)
data['max_power'] = data['max_power'].str.extract(r'(\d+\.\d+)').astype(float)

# Encoding all non-numeric columns
data = data.apply(lambda col: LabelEncoder().fit_transform(col) if col.dtypes == 'object' else col)

# Drop remaining NaNs
data = data.dropna()

X = data[['year', 'km_driven', 'fuel', 'seller_type', 'transmission', 'owner', 'mileage', 'engine', 'max_power']]
y_reg = data['selling_price']
y_clf = (y_reg > y_reg.median()).astype(int)

X_train, X_test, y_reg_train, y_reg_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)
_, _, y_clf_train, y_clf_test = train_test_split(X, y_clf, test_size=0.2, random_state=42)

# Function to display new window with results
def show_results(title, results):
    result_window = Toplevel(root)
    result_window.title(title)

    Label(result_window, text=results, font=("Arial", 12), justify="left", wraplength=500).pack(pady=10)

# Function to display new window with plots
def show_plot_window(title, plot_function):
    plot_window = Toplevel(root)
    plot_window.title(title)

    # Set default window size
    plot_window.geometry("800x600")

    # Create the figure
    fig, ax = plt.subplots(figsize=(8, 6))  # Adjust figure size to fit window
    plot_function(ax)
    fig.tight_layout()

    # Embed the plot in Tkinter
    canvas = FigureCanvasTkAgg(fig, master=plot_window)

    # Add toolbar on top
    toolbar = NavigationToolbar2Tk(canvas, plot_window)
    toolbar.update()
    toolbar.pack(side=tk.TOP, fill=tk.X)

    # Add canvas below the toolbar
    canvas_widget = canvas.get_tk_widget()
    canvas_widget.pack(fill=tk.BOTH, expand=True)

    canvas.draw()

# EDA Functions
def eda_window():
    eda_win = Toplevel(root)
    eda_win.title("Exploratory Data Analysis")

    def show_dataset():
        dataset_window = Toplevel(eda_win)
        dataset_window.title("Dataset Overview")
        text_area = scrolledtext.ScrolledText(dataset_window, wrap=tk.WORD, width=100, height=30, font=("Arial", 10))
        text_area.pack(pady=10, padx=10)
        text_area.insert(tk.INSERT, data.head(20).to_string())

    def show_heatmap():
        def plot_heatmap(ax):
            correlation_matrix = data.corr()
            sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, ax=ax)
            ax.set_title("Heatmap of Correlations")

        show_plot_window("Heatmap of Dataset", plot_heatmap)

    def correlation():
        def plot_correlation():
            selected_features = ['owner', 'km_driven', 'selling_price', 'fuel']
            pairplot_data = data[selected_features]
            pairplot = sns.pairplot(pairplot_data, hue='fuel', diag_kind='kde', markers=["o", "s", "D", "^"])
            pairplot.fig.tight_layout()
            canvas = FigureCanvasTkAgg(pairplot.fig, master=plot_window)
            canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

        plot_window = Toplevel(root)
        plot_window.title("Correlation")
        plot_correlation()

    def selling_price_count():
        def plot_count(ax):
            sns.set_style('darkgrid')
            sns.histplot(data=data, x='selling_price', bins=100, kde=True, edgecolor='none', alpha=0.4, line_kws={'linestyle': '--', 'lw': 2.5}, ax=ax)
            ax.set_title("Selling Price Count")
            ax.set_xlabel("Selling Price")
            ax.set_ylabel("Count")

        show_plot_window("Selling Price Count", plot_count)

    Button(eda_win, text="Show Dataset", command=show_dataset).pack(pady=5)
    Button(eda_win, text="Show Heatmap", command=show_heatmap).pack(pady=5)
    Button(eda_win, text="Correlation", command=correlation).pack(pady=5)
    Button(eda_win, text="Selling Price Count", command=selling_price_count).pack(pady=5)

lin_reg = LinearRegression()
lin_reg.fit(X_train[['year', 'km_driven', 'mileage', 'engine', 'max_power']], y_reg_train)

def linear_regression_window():
    def predict():
        try:
            # Ambil nilai input dari formulir
            year = float(year_entry.get())
            km_driven = float(km_driven_entry.get())
            mileage = float(mileage_entry.get())
            engine = float(engine_entry.get())
            max_power = float(max_power_entry.get())

            # Siapkan data untuk prediksi
            input_data = pd.DataFrame([[year, km_driven, mileage, engine, max_power]],
                                      columns=['year', 'km_driven', 'mileage', 'engine', 'max_power'])
            
            # Lakukan prediksi
            prediction = lin_reg.predict(input_data)[0]
            
            # Hitung MSE
            y_pred = lin_reg.predict(X_test[['year', 'km_driven', 'mileage', 'engine', 'max_power']])
            mse = mean_squared_error(y_reg_test, y_pred)
            
            # Tampilkan hasil
            result_label.config(text=f"Predicted Selling Price: {prediction:.2f}\nMSE: {mse:.2f}")
        except Exception as e:
            result_label.config(text=f"Error: {e}")

    def fill_random():
        # Isi input dengan nilai acak
        year_entry.delete(0, tk.END)
        km_driven_entry.delete(0, tk.END)
        mileage_entry.delete(0, tk.END)
        engine_entry.delete(0, tk.END)
        max_power_entry.delete(0, tk.END)

        year_entry.insert(0, str(random.randint(2000, 2022)))
        km_driven_entry.insert(0, str(random.randint(10000, 200000)))
        mileage_entry.insert(0, f"{random.uniform(10, 25):.1f}")
        engine_entry.insert(0, str(random.randint(1000, 3000)))
        max_power_entry.insert(0, str(random.randint(50, 200)))

    # Buat jendela baru untuk prediksi Linear Regression
    lr_window = Toplevel(root)
    lr_window.title("Linear Regression Prediction")
    lr_window.geometry("400x500")

    # Judul
    Label(lr_window, text="Linear Regression Prediction", font=("Arial", 16)).pack(pady=10)

    # Formulir input
    Label(lr_window, text="Year of Manufacture:").pack(pady=5)
    year_entry = Entry(lr_window)
    year_entry.pack()

    Label(lr_window, text="Kilometers Driven:").pack(pady=5)
    km_driven_entry = Entry(lr_window)
    km_driven_entry.pack()

    Label(lr_window, text="Mileage (km/l):").pack(pady=5)
    mileage_entry = Entry(lr_window)
    mileage_entry.pack()

    Label(lr_window, text="Engine Capacity (cc):").pack(pady=5)
    engine_entry = Entry(lr_window)
    engine_entry.pack()

    Label(lr_window, text="Maximum Power (hp):").pack(pady=5)
    max_power_entry = Entry(lr_window)
    max_power_entry.pack()

    # Tombol untuk prediksi dan random data
    Button(lr_window, text="Predict Selling Price", command=predict, bg="#007bff", fg="white").pack(pady=10)
    Button(lr_window, text="Random", command=fill_random, bg="#007bff", fg="white").pack(pady=5)

    # Label hasil
    result_label = Label(lr_window, text="", font=("Arial", 12), fg="green")
    result_label.pack(pady=10)

# SVR Function
def svr_regression():
    def run_svr(kernel, C, gamma):
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        svr = SVR(kernel=kernel, C=C, gamma=gamma)
        svr.fit(X_train_scaled, y_reg_train)
        y_pred = svr.predict(X_test_scaled)
        mse = mean_squared_error(y_reg_test, y_pred)

        results = f"SVR with kernel={kernel}, C={C}, gamma={gamma}\nMSE: {mse:.4f}"
        show_results("SVR Results", results)

    svr_window = Toplevel(root)
    svr_window.title("SVR Parameters")

    Label(svr_window, text="C:").grid(row=0, column=0)
    c_entry = Entry(svr_window)
    c_entry.grid(row=0, column=1)

    kernel_var = StringVar(value="linear")
    Label(svr_window, text="Kernel:").grid(row=1, column=0)
    Radiobutton(svr_window, text="Linear", variable=kernel_var, value="linear").grid(row=1, column=1)
    Radiobutton(svr_window, text="Poly", variable=kernel_var, value="poly").grid(row=1, column=2)
    Radiobutton(svr_window, text="RBF", variable=kernel_var, value="rbf").grid(row=1, column=3)

    gamma_var = StringVar(value="scale")
    Label(svr_window, text="Gamma:").grid(row=2, column=0)
    Radiobutton(svr_window, text="Scale", variable=gamma_var, value="scale").grid(row=2, column=1)
    Radiobutton(svr_window, text="Auto", variable=gamma_var, value="auto").grid(row=2, column=2)

    Button(svr_window, text="Run SVR", command=lambda: run_svr(
        kernel_var.get(),
        float(c_entry.get()),
        gamma_var.get()
    )).grid(row=3, columnspan=4, pady=10)

# KNN Classification Function
def knn_classification():
    def run_knn(n_neighbors, weights):
        knn = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
        knn.fit(X_train, y_clf_train)
        y_pred = knn.predict(X_test)

        report = classification_report(y_clf_test, y_pred)
        show_results("KNN Results", report)

    knn_window = Toplevel(root)
    knn_window.title("KNN Parameters")

    Label(knn_window, text="Neighbors:").grid(row=0, column=0)
    neighbors_entry = Entry(knn_window)
    neighbors_entry.grid(row=0, column=1)

    weights_var = StringVar(value="uniform")
    Label(knn_window, text="Weights:").grid(row=1, column=0)
    Radiobutton(knn_window, text="Uniform", variable=weights_var, value="uniform").grid(row=1, column=1)
    Radiobutton(knn_window, text="Distance", variable=weights_var, value="distance").grid(row=1, column=2)

    Button(knn_window, text="Run KNN", command=lambda: run_knn(
        int(neighbors_entry.get()),
        weights_var.get()
    )).grid(row=2, columnspan=3, pady=10)

# Decision Tree Function
def decision_tree():
    def run_tree(max_depth, min_samples_split):
        dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
        dt.fit(X_train, y_clf_train)
        y_pred = dt.predict(X_test)

        report = classification_report(y_clf_test, y_pred)

        def plot_tree_func(ax):
            plot_tree(dt, feature_names=X.columns, class_names=['Low', 'High'], filled=True, ax=ax)

        show_results("Decision Tree Results", report)
        show_plot_window("Decision Tree Visualization", plot_tree_func)

    tree_window = Toplevel(root)
    tree_window.title("Decision Tree Parameters")

    Label(tree_window, text="Max Depth:").grid(row=0, column=0)
    depth_entry = Entry(tree_window)
    depth_entry.grid(row=0, column=1)

    Label(tree_window, text="Min Samples Split:").grid(row=1, column=0)
    split_entry = Entry(tree_window)
    split_entry.grid(row=1, column=1)

    Button(tree_window, text="Run Tree", command=lambda: run_tree(
        int(depth_entry.get()),
        int(split_entry.get())
    )).grid(row=2, columnspan=2, pady=10)

# K-Means Clustering Function
def kmeans_clustering():
    def run_kmeans(n_clusters):
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        kmeans.fit(X)
        labels = kmeans.labels_

        def plot_kmeans_func(ax):
            ax.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')
            ax.set_title("K-Means Clustering")
            ax.set_xlabel("Feature 1")
            ax.set_ylabel("Feature 2")

        results = f"K-Means Clustering Completed with {n_clusters} Clusters."
        show_results("K-Means Results", results)
        show_plot_window("K-Means Visualization", plot_kmeans_func)

    kmeans_window = Toplevel(root)
    kmeans_window.title("K-Means Parameters")

    Label(kmeans_window, text="Number of Clusters:").grid(row=0, column=0)
    clusters_entry = Entry(kmeans_window)
    clusters_entry.grid(row=0, column=1)

    Button(kmeans_window, text="Run K-Means", command=lambda: run_kmeans(
        int(clusters_entry.get())
    )).grid(row=1, columnspan=2, pady=10)

# Main Application
root = tk.Tk()
root.title("Machine Learning Toolkit")

Button(root, text="Exploratory Data Analysis (EDA)", command=eda_window).pack(pady=5)
Button(root, text="Linear Regression", command=linear_regression_window).pack(pady=5)
Button(root, text="SVR Regression", command=svr_regression).pack(pady=5)
Button(root, text="KNN Classification", command=knn_classification).pack(pady=5)
Button(root, text="Decision Tree Classification", command=decision_tree).pack(pady=5)
Button(root, text="K-Means Clustering", command=kmeans_clustering).pack(pady=5)
Button(root, text="Exit", command=root.quit).pack(pady=5)

root.mainloop()
```
